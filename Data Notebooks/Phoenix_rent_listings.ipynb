```python
import requests

url = "https://realty-mole-property-api.p.rapidapi.com/rentalListings"

querystring = {"city":"Phoenix","state":"AZ","limit":"1000"}

headers = {
	"X-RapidAPI-Key": "{key}",
	"X-RapidAPI-Host": "realty-mole-property-api.p.rapidapi.com"
}

response_rent = requests.request("GET", url, headers=headers, params=querystring)

print(response_rent.text)
```

```python
import pandas as pd
 
response_rent_df = pd.read_json(response_rent.text)
```

```python
response_rent_df.to_csv('Phoenix_rent_listings.csv')
```
