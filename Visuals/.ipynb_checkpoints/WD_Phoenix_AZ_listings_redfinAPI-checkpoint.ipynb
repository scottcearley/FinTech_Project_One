```python
# importing libraries
import requests
import json 
import pandas as pd
import hvplot.pandas

# pulling API data from RedFin
url = "https://unofficial-redfin.p.rapidapi.com/properties/list"

querystring = {"region_id":"14240","region_type":"6","uipt":"1,2,3,7,8","status":"9","sf":"1,2,5,6,7","num_homes":"300","sold_within_days":"30"}

headers = {
	"X-RapidAPI-Key": "b6a21fcaa0msh2e9de61afa99433p14003ajsnd33bf08e42ad",
	"X-RapidAPI-Host": "unofficial-redfin.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers, params=querystring)

data = response.json()

# display API pull status
print(f"Start API acquisition")
print(f"------------&&&&-------------------")

print(json.dumps(data, indent=4))

```

```python
# identifying column names for dataframe to be created
property_id = []
property_type = []
beds = []
baths = []
price_info = []
square_foot = []
#sale_date = ''
#price_per_sf = []
latitude = [] 
longitude = []

# for loop to extract variables from JSON file
for homeData in data['homes']:
    property_id.append(homeData['homeData']['propertyId'])
    property_type.append(homeData['homeData']['propertyType'])
    #beds.append(homeData['homeData']['beds']['value'])
    #baths.append(homeData['homeData']['baths']['value'])
    price_info.append(homeData['homeData']['priceInfo']['amount']['value'])
    square_foot.append(homeData['homeData']['sqftInfo']['amount']['value'])
    #sale_date.append(homeData['homeData']['sashes']['lastSaleDate'])
    latitude.append(homeData['homeData']['addressInfo']['centroid']['centroid']['latitude'])
    longitude.append(homeData['homeData']['addressInfo']['centroid']['centroid']['longitude'])

```

```python
#creating dataframe
data_dict = {"Property_ID": property_id, "Property_Type": property_type, "Price Info": price_info, "Square Foot": square_foot, "Latitude": latitude, "Longitude": longitude}
data_df = pd.DataFrame(data_dict)

data_df['Price Info']=data_df['Price Info'].astype(float)
data_df['Square Foot']=data_df['Square Foot'].astype(float)

data_df
```

```python
data_df.hvplot.points(
    'Longitude',
    'Latitude', 
    geo = True, 
    frame_width = 700, 
    frame_height = 500, 
    tiles = 'OSM',
    color = 'Property_Type',
    #size = 'Square Foot',
    hover_cols = 'Price Info',
    title = 'Phoenix Housing Property Listings'
)
```

```python
price_per_sf = data_df['Price Info'] / data_df['Square Foot']
print(price_per_sf)
```
